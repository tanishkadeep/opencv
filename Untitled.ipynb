```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
original = cv.imread('mountain.jpg')

# Convert the image to grayscale
gray_im = cv.cvtColor(original, cv.COLOR_BGR2GRAY)

# Correct the grayscale image by applying a gamma correction
gray_correct = np.array(255 * (gray_im / 255) ** 1.2, dtype='uint8')

# Perform histogram equalization
gray_equ = cv.equalizeHist(gray_im)

# Apply adaptive thresholding to segment the image
thresh = cv.adaptiveThreshold(gray_correct, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresh = cv.bitwise_not(thresh)

# Erosion and dilation to connect regions
kernel = np.ones((15, 15), np.uint8)
img_dilation = cv.dilate(thresh, kernel, iterations=1)
img_erode = cv.erode(img_dilation, kernel, iterations=1)

# Apply median blur to clean up noise
img_erode = cv.medianBlur(img_erode, 7)

# Label connected components (regions)
ret, labels = cv.connectedComponents(img_erode)

# Generate a color map for visualization of labeled regions
label_hue = np.uint8(179 * labels / np.max(labels))
blank_ch = 255 * np.ones_like(label_hue)
labeled_img = cv.merge([label_hue, blank_ch, blank_ch])
labeled_img = cv.cvtColor(labeled_img, cv.COLOR_HSV2BGR)

# Set background label to black
labeled_img[label_hue == 0] = 0

# Show the final labeled image with object count
plt.subplot(1, 2, 2)
plt.title(f'Objects Counted: {ret - 1}')
plt.imshow(labeled_img)

# Print the total number of objects
print(f'Objects number is: {ret - 1}')

# Show the plot
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load and process image
img = cv.imread('f1.png', 0)  # Read in grayscale directly
img = np.array(255 * (img / 255) ** 1.2, dtype='uint8')  # Apply gamma correction
img = cv.equalizeHist(img)  # Apply histogram equalization

thresh = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
thresh = cv.bitwise_not(thresh)  # Invert binary image

# Morphological operations & noise cleanup
kernel = np.ones((15, 15), np.uint8)
img = cv.medianBlur(cv.erode(cv.dilate(thresh, kernel), kernel), 7)

# Connected components and labeling
ret, labels = cv.connectedComponents(img)
labeled_img = cv.applyColorMap(np.uint8(200 * labels / labels.max()), cv.COLORMAP_JET)
labeled_img[labels == 0] = 0  # Set background to black

# Display results
plt.subplot(1, 2, 2)
plt.title(f'Objects Counted: {ret - 1}')
plt.imshow(labeled_img)
print(f'Objects number is: {ret - 1}')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('f1.png')

# Convert image to RGB (OpenCV loads images in BGR format by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply Gaussian Blur with a 15x15 kernel
filtered_image = cv2.GaussianBlur(image, (15, 15), 0)

# Convert the filtered image to RGB for display
filtered_image_rgb = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB)

# Plot the original and filtered images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Filtered Image
plt.subplot(1, 2, 2)
plt.imshow(filtered_image_rgb)
plt.title('Gaussian Filtered Image')
plt.axis('off')

# Adjust layout and display the images
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load images and convert to grayscale
image1 = cv2.imread('I1.png')
image2 = cv2.imread('I2.png')
gray1, gray2 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY), cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Detect SIFT features and match
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(gray1, None)
kp2, des2 = sift.detectAndCompute(gray2, None)
matches = cv2.BFMatcher().knnMatch(des1, des2, k=2)
good_matches = [m for m, n in matches if m.distance < 0.75 * n.distance]

# Find homography and warp image
src_pts = np.float32([kp1[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
dst_pts = np.float32([kp2[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
H, _ = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
result = cv2.warpPerspective(image1, H, (image2.shape[1] + image1.shape[1], image2.shape[0]))
result[0:image2.shape[0], 0:image2.shape[1]] = image2

# Display images
for i, img in enumerate([image1, image2, result], 1):
    plt.subplot(1, 3, i)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
plt.show()
```
